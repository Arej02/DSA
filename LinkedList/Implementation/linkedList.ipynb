```python
class Node:

    def __init__(self,data):
        self.data=data
        self.next=None


class LinkedList:

    def __init__(self):
        self.head=None
        self.n=0

    def __len__(self):
        return self.n

    def insert_head(self,value):
        new_node=Node(value)
        new_node.next=self.head
        self.head=new_node
        self.n+=1

    def traverse(self):
        temp=self.head
        values=[]
        
        while temp!=None:
            values.append(str(temp.data))
            temp=temp.next

        return "->".join(values)

    def insert_tail(self,value):
        new_node=Node(value)
        temp=self.head

        while temp.next!=None:
            temp=temp.next

        temp.next=new_node




```

```python
ll1=LinkedList()
```

```python
ll1.insert_head(1)
ll1.insert_head(2)
ll1.insert_head(3)
ll1.insert_head(4)
```

```python
ll1.traverse()
```

```python
ll1.insert_tail(45)
```
