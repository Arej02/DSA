```python
class Node:

    def __init__(self,data):
        self.data=data
        self.next=None


class LinkedList:

    def __init__(self):
        self.head=None
        self.n=0

    # Length of the Linked List:
    def __len__(self):
        return self.n

    # Insert a new head:
    def insert_head(self,value):
        new_node=Node(value)
        new_node.next=self.head
        self.head=new_node
        self.n+=1

    # Traverse the linked list:
    def traverse(self):
        temp=self.head
        values=[]
        
        while temp!=None:
            values.append(str(temp.data))
            temp=temp.next

        return "->".join(values)

    # Insert after the tail:
    def insert_tail(self,value):
        new_node=Node(value)
        temp=self.head

        while temp.next!=None:
            temp=temp.next

        temp.next=new_node
        self.n+=1

    # Insert in the middle:
    def insert_middle(self,value):
        new_node=Node(value)
        temp=self.head
        curr=self.head

        for _ in range(self.n//2):
            curr=temp
            temp=temp.next

        curr.next=new_node
        new_node.next=temp
        self.n+=1

    def clear(self):
        self.head=None
        self.n=0

    # Delete the head:
    def delete_head(self):
        
        if self.n==0 or self.head is None:
            print("No LL")
            return

        if self.n==1:
            print("The LL deleted")

        self.head=self.head.next
        self.n-=1

    # Delete the tail:
    def delete_tail(self):
        temp=self.head

        while temp.next.next != None:
            temp=temp.next

        temp.next=None
        self.n-=1

    # Delete by the value:
    def delete_val(self,val):
        current=self.head
        previous=self.head

        while current.data!=val:
            previous=current
            current=current.next

        previous.next=current.next
        current.next=None
        
        self.n-=1

    # Search by value:
    def search_value(self,val):
        temp=self.head
        position=0

        while temp!=None:

            if temp.data==val:
                return val
            else:
                temp=temp.next
                position+=1

        return "The data is not present"
    
```
